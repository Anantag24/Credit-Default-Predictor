```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, confusion_matrix,
                           roc_auc_score, classification_report,
                           balanced_accuracy_score, precision_recall_curve,
                           roc_curve)  # Added roc_curve here
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns

# ======================
# 1. DATA PREPARATION
# ======================
def load_or_generate_data(data_path=None, n_samples=10000):
    """
    Load real data if available, else generate realistic synthetic data
    """
    if data_path:
        try:
            df = pd.read_csv(data_path)
            print("‚úÖ Real data loaded successfully")
        except:
            print("‚ö†Ô∏è Couldn't load real data - generating synthetic data")
            df = generate_realistic_data(n_samples)
    else:
        print("üî∑ Generating synthetic data")
        df = generate_realistic_data(n_samples)

    # Basic validation
    assert 'default' in df.columns, "Data must contain 'default' column"
    assert len(df) > 1000, "Dataset too small (<1000 samples)"

    return df

def generate_realistic_data(n_samples=10000):
    """Generate data with realistic financial patterns"""
    np.random.seed(42)

    # Base features
    age = np.random.normal(45, 15, n_samples).clip(18, 70).astype(int)
    income = np.abs(np.random.lognormal(10.5, 0.4, n_samples)).astype(int)
    credit_score = np.random.normal(650, 100, n_samples).clip(300, 850).astype(int)

    # Realistic default probability (5-15% default rate)
    default_prob = (
        0.3 * (credit_score < 580) +
        0.2 * (income < 30000) +
        0.1 * (age > 65) +
        0.05 * np.random.random(n_samples)
    )
    default = (default_prob > np.percentile(default_prob, 85)).astype(int)  # ~15% defaults

    # Financial features with noise
    loan_amount = (income * np.random.uniform(0.1, 0.8, n_samples)).astype(int)
    current_debt = (income * np.random.uniform(0.05, 0.6, n_samples)).astype(int)

    return pd.DataFrame({
        'age': age,
        'income': income,
        'credit_score': credit_score,
        'loan_amount': loan_amount,
        'current_debt': current_debt,
        'employment_length': np.random.randint(0, 10, n_samples),  # Years
        'num_credit_lines': np.random.randint(1, 10, n_samples),
        'default': default
    })

# ======================
# 2. FEATURE ENGINEERING
# ======================
def create_features(df):
    """Create meaningful financial ratios"""
    df = df.copy()

    # Financial ratios
    df['debt_to_income'] = df['current_debt'] / (df['income'] + 1)
    df['loan_to_income'] = df['loan_amount'] / (df['income'] + 1)
    df['credit_utilization'] = df['current_debt'] / (df['loan_amount'] + 1)

    # Binning continuous variables
    df['credit_score_bin'] = pd.cut(df['credit_score'],
                                   bins=[300, 580, 670, 740, 800, 850],
                                   labels=['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'])

    # One-hot encoding
    df = pd.get_dummies(df, columns=['credit_score_bin'], drop_first=True)

    return df.dropna()

# ======================
# 3. MODEL PIPELINE
# ======================
def build_model_pipeline(X_train, y_train):
    """Train multiple models with class balancing"""
    # Balance classes using SMOTE
    smote = SMOTE(random_state=42)
    X_res, y_res = smote.fit_resample(X_train, y_train)

    models = {
        'Logistic Regression': LogisticRegression(class_weight='balanced', max_iter=1000),
        'Random Forest': RandomForestClassifier(class_weight='balanced_subsample',
                                              max_depth=5,
                                              n_estimators=100)
    }

    trained_models = {}
    for name, model in models.items():
        model.fit(X_res, y_res)
        trained_models[name] = model
        print(f"‚úÖ {name} trained")

    return trained_models

# ======================
# 4. EVALUATION
# ======================
def evaluate_model(model, X_test, y_test):
    """Comprehensive model evaluation"""
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    print("\nüìä Classification Report:")
    print(classification_report(y_test, y_pred, target_names=["Safe", "Risky"]))

    print("Balanced Accuracy:", balanced_accuracy_score(y_test, y_pred))
    print("AUC-ROC:", roc_auc_score(y_test, y_proba))

    # Confusion Matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix(y_test, y_pred),
                annot=True, fmt='d', cmap='Blues',
                xticklabels=['Predicted Safe', 'Predicted Risky'],
                yticklabels=['Actual Safe', 'Actual Risky'])
    plt.title('Confusion Matrix')
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc_score(y_test, y_proba):.2f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show()

# ======================
# MAIN WORKFLOW
# ======================
def main(data_path=None):
    print("=== CREDIT RISK MODELING SYSTEM ===")

    # 1. Data Loading
    print("\n[1/4] Loading/generating data...")
    df = load_or_generate_data(data_path)
    print(f"Dataset shape: {df.shape}")
    print(f"Default rate: {df['default'].mean():.2%}")

    # 2. Feature Engineering
    print("\n[2/4] Creating features...")
    processed_df = create_features(df)
    X = processed_df.drop('default', axis=1)
    y = processed_df['default']

    # 3. Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    # 4. Feature Scaling
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # 5. Model Training
    print("\n[3/4] Training models...")
    models = build_model_pipeline(X_train, y_train)

    # 6. Evaluation
    print("\n[4/4] Evaluating models...")
    for name, model in models.items():
        print(f"\nüîç {name} Performance:")
        evaluate_model(model, X_test, y_test)

        # Cross-validation
        cv_scores = cross_val_score(model, X_train, y_train,
                                  cv=5, scoring='roc_auc')
        print(f"Cross-validated AUC-ROC: {np.mean(cv_scores):.3f} ¬± {np.std(cv_scores):.3f}")

if __name__ == "__main__":
    # Usage: Pass path to your data or None for synthetic data
    main(data_path=None)  # Set data_path="your_data.csv" for real data
```
